````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots(figsize=(6, 3))
p1, p2, p3 = ax.plot(x, x, label='linear'), ax.plot(x, x**3, label='cubic'), ax.plot(x, x**4, label='quadratic')
ax.legend(loc='upper left')
ax.set_xlabel('x label'), ax.set_ylabel('y label'), ax.set_title('Simple Plot')
plt.show()
````
