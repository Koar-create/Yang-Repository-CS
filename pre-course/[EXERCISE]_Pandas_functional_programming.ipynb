---
# Q1
---

```python
A = df.groupby(['Gender', 'City_Category']).mean()
A
print(f"The mean purchase of women in the city A is {A.loc['F'].loc['A'].Purchase:.2f}. ")
```

# Q2
---

```python
B = df.groupby(['Gender', 'City_Category']).count()
B
print(f"The total number of men completed purchase in the city B is {B.loc['M'].loc['B'].Purchase}. ")
```
